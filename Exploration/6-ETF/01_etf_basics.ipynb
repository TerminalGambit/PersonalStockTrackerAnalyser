## üß† What is an ETF?

An **ETF (Exchange-Traded Fund)** is a basket of securities‚Äîstocks, bonds, commodities‚Äîthat you can buy or sell on a stock exchange, just like a regular stock.

### üîç Key Characteristics
- Trades like a stock during market hours.
- Holds a diversified collection of assets.
- Typically follows an index or sector.
- Lower fees than mutual funds.

### üÜö ETF vs Mutual Fund
| Feature | ETF | Mutual Fund |
|--------|-----|-------------|
| Traded on exchange | ‚úÖ | ‚ùå |
| Intra-day pricing | ‚úÖ | ‚ùå |
| Passive investment (often) | ‚úÖ | ‚ùå |
| Expense ratio | Low | Medium |


## üß¨ Types of ETFs

### üìä Index ETFs
Track a specific index (e.g., S&P 500).
- `SPY` ‚Äì S&P 500 ETF
- `VOO` ‚Äì Vanguard S&P 500 ETF

### üíº Sector ETFs
Target specific sectors:
- `XLK` ‚Äì Technology
- `XLF` ‚Äì Financials
- `XLE` ‚Äì Energy

### üíµ Bond ETFs
- `BND` ‚Äì Total Bond Market
- `TLT` ‚Äì 20+ Year Treasuries

### ü™ô Commodity ETFs
- `GLD` ‚Äì Gold
- `SLV` ‚Äì Silver

### üöÄ Leveraged & Inverse ETFs
- `TQQQ`, `SQQQ` ‚Äì 3x Nasdaq
- ‚ö†Ô∏è Risky and not for long-term holding


## üßÆ How to Choose an ETF

### üìå Consider:
- **Expense Ratio**: Annual fee as % of investment.
- **Liquidity**: Volume traded daily.
- **Assets Under Management (AUM)**: Larger often = safer.
- **Tracking Error**: Difference between ETF and index.
- **Holdings Transparency**
- **Dividend Yield** & tax treatment


## üìà Market Trends in ETFs

- Massive rise in **passive investing** via ETFs.
- **Vanguard**, **iShares**, and **SPDR** dominate ETF flows.
- Thematic ETFs (AI, blockchain, EVs) gaining popularity.
- Global ETF assets exceed $10 trillion (2024 est).


```python
import yfinance as yf

etfs = {
    "S&P 500": "SPY",
    "Nasdaq 100": "QQQ",
    "Technology": "XLK",
    "Bond Market": "BND",
    "Gold": "GLD"
}

data = {name: yf.Ticker(ticker) for name, ticker in etfs.items()}
```

```python
for name, etf in data.items():
    info = etf.info
    print(f"üìò {name} ({etf.ticker}):")
    print(f" - Long Name: {info.get('longName', 'N/A')}")
    print(f" - Category: {info.get('category', 'N/A')}")
    print(f" - Expense Ratio: {info.get('expenseRatio', 'N/A')}")
    print(f" - Total Assets: {info.get('totalAssets', 'N/A')}")
    print(f" - NAV: {info.get('navPrice', 'N/A')}")
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
for name, etf in data.items():
    hist = etf.history(period='6mo')
    plt.plot(hist['Close'], label=name)

plt.title("ETF Price Trends (6 Months)")
plt.ylabel("Price (USD)")
plt.xlabel("Date")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
