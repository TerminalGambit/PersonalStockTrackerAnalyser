# ðŸ“˜ Introduction to ETFs (Exchange-Traded Funds)
An **ETF** is a type of investment fund that is traded on stock exchanges, much like stocks.
They typically aim to track the performance of a specific index (e.g., S&P 500) or sector (e.g., tech).

âœ… **Main Features:**
- Traded like a stock
- Typically passive (index-tracking)
- Low fees
- Diversification with a single purchase


```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## ðŸŽ¯ Let's analyze an example ETF: **SPY**
`SPY` tracks the S&P 500 index. We'll fetch its historical data and explore basic features.

```python
etf = yf.Ticker('SPY')
df = etf.history(period='6mo')
df.head()
```

## ðŸ“ˆ Price Chart

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Close'], label='SPY Close Price')
plt.title('SPY ETF - Closing Price (Last 6 Months)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

## ðŸ” Daily Returns

```python
df['Daily Return'] = df['Close'].pct_change()
df['Daily Return'].plot(figsize=(12, 4), title='SPY Daily Returns')
```
