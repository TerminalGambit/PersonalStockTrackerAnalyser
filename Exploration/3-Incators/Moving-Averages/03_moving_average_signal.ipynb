This notebook explores the 50-day Moving Average (50MA) for a chosen stock using Yahoo Finance data. It helps answer the question: **"Should I buy this stock?"** based on trend-following signals.


```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 1: Choose a Stock and Load Historical Data

We will use the yfinance library to get 1 year of historical price data for a stock (default: Microsoft - MSFT).


```python
ticker = yf.Ticker("MSFT")
df = ticker.history(period="1y")
df.head()
```

## Step 2: Calculate the 50-Day Moving Average

This smooths out the daily price fluctuations and helps identify overall trends.


```python
df["50MA"] = df["Close"].rolling(window=50).mean()
df[["Close", "50MA"]].tail()
```

## Step 3: Visualize the Price and Moving Average

We'll plot both the actual closing price and the 50-day moving average.


```python
plt.figure(figsize=(14, 6))
plt.plot(df["Close"], label="Close Price", linewidth=2)
plt.plot(df["50MA"], label="50-Day Moving Average", linestyle="--")
plt.title("MSFT Close Price & 50-Day MA")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

## Step 4: Interpretation

- When the stock price is **above the 50MA**, it's often seen as an **uptrend** (bullish).
- When the price is **below the 50MA**, it may suggest a **downtrend** (bearish).
- Traders look for **crossovers** where the price moves through the MA line to spot buy/sell signals.
